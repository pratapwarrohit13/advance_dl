```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(features_train, target_train), (features_test, target_test) = mnist.load_data()
```

```python
unique, counts = np.unique(target_train, return_counts=True)
print("Train labels: ", dict(zip(unique, counts)))
```

```python
unique, counts = np.unique(target_test, return_counts=True)
print("Test labels: ", dict(zip(unique, counts)))
```

```python
indexes = np.random.randint(0, features_train.shape[0], size=25)
images = features_train[indexes]
labels = target_train[indexes]
```

```python
plt.figure(figsize=(5,5))
for i in range(len(indexes)):
    plt.subplot(5, 5, i + 1)
    image = images[i]
    plt.imshow(image, cmap='gray')
    plt.axis('off')

plt.savefig("mnist-samples.png")
plt.show()
plt.close('all')
```
