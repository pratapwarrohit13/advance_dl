```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, SimpleRNN
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras.datasets import mnist
```

```python
(features_train, target_train), (features_test, target_test) = mnist.load_data()
```

```python
num_labels = len(np.unique(target_train))
```

```python
target_train = to_categorical(target_train)
target_test = to_categorical(target_test)
```

```python
image_size = features_train.shape[1]
features_train= np.reshape(features_train,[-1, image_size, image_size])
features_test = np.reshape(features_test,[-1, image_size, image_size])
features_train = features_train.astype('float32') / 255
features_test = features_test.astype('float32') / 255
```

```python
input_shape = (image_size, image_size)
batch_size = 128
units = 256
dropout = 0.2
```

```python
model = Sequential()
model.add(SimpleRNN(units=units,
                    dropout=dropout,
                    input_shape=input_shape))
model.add(Dense(num_labels))
model.add(Activation('softmax'))
model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
model.fit(features_train, target_train, epochs=50, batch_size=batch_size)
```

```python
_, acc = model.evaluate(features_test,
                        target_test,
                        batch_size=batch_size,
                        verbose=0)
print("\nTest accuracy: %.1f%%" % (100.0 * acc))
```
